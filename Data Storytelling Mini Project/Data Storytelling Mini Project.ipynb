# Data Storytelling Mini Project

This notebook works with the "Effect of COVID19 on Trade" dataset, which can be found here: https://www.kaggle.com/kkhandekar/effects-of-covid19-on-trade

This dataset shows weekly updates on the trade between New Zealand and various countries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("covid19nztrade.csv.zip")
```

```python
df
```

```python
# some of the trade is measured in tonnes, we will isolate that which is measured in $
df_dollars = df[df['Measure'] == '$'].drop(columns='Measure')
df = df_dollars
```

```python
# for simplicity's sake, let's focus only on how covid has affected NZ's exports
df_exports = df[df['Direction']=='Exports'].drop(columns='Direction')
df = df_exports
df
```
